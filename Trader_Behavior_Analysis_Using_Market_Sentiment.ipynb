```python
import gdown
import os

dataset = "dataset"
os.makedirs(dataset, exist_ok=True)

url = "https://drive.google.com/uc?id=1PgQC0tO8XN-wqkNyghWc_-mnrYv_nhSf"

output = "dataset/Fear_Greed_Index.csv"

gdown.download(url, output, quiet=False)
```

```python
sentiment = pd.read_csv("dataset/Fear_Greed_Index.csv")

sentiment.head()
```

```python
sentiment.info()
```

```python
sentiment["date"] = pd.to_datetime(sentiment["date"])
sentiment["date"] = sentiment["date"].dt.date
```

```python
data = pd.merge(trader , sentiment, on="date" , how="left")
data.head()
```

```python
data["classification"].value_counts()
```

```python
data.isnull().sum()
```

```python
data["profit"] = data["Closed PnL"] > 0

data["trade_value"] = data["Execution Price"] * data["Size Tokens"]

data["hour"] = data["Timestamp IST"].dt.hour
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x="hour", data=data)

plt.title("Trading Activity by Hour")

plt.show()
```

```python
sns.boxplot(x="classification", y="Closed PnL", data=data)

plt.title("Profit Distribution by Market Sentiment")

plt.xticks(rotation=45)

plt.show()
```

```python
sns.boxplot(x="classification", y="Size USD", data=data)

plt.title("Trade Size Distribution by Market Sentiment")

plt.xticks(rotation=45)

plt.show()
```

```python
data.isnull().sum()
```

```python
data = data.dropna(subset=["classification"])
```

```python
data['value'].fillna(data['value'].mean(), inplace=True)
```

```python
data['Timestamp'].fillna(data['Timestamp'].mode()[0], inplace=True)
```

```python
order = ["Extreme Fear","Fear","Neutral","Greed","Extreme Greed"]

sns.boxplot(x="classification", y="Closed PnL", data=data, order=order)
```

```python
numeric = data[["Closed PnL","Size USD","trade_value","profit"]]

sns.heatmap(numeric.corr(), annot=True)

plt.title("Correlation Between Trading Metrics")

plt.show()
```

**Key Insights**

Insight 1

Traders achieve the highest win rate during Extreme Greed sentiment, indicating strong bullish momentum improves trading outcomes.

Insight 2

Extreme Fear sentiment results in lower profitability due to panic-driven trading behavior.

Insight 3

Trade sizes tend to increase during Greed markets, suggesting higher risk appetite among traders.

Insight 4

Trading activity peaks during specific hours, likely corresponding to periods of high market liquidity.
